Проверка гипотезы: Как критики относятся к спортивным играм?
*  Н0 - Критики ставят спортивным играм более высокие оценки
*  Н1 - Оценки критиков спортивным играм не отличаются от оценок игр других жанров

Сначала проверим данные на нормальность распределения с помощью теста Шапиро-Уилка

```python
import scipy.stats as st
```

```python
df_score = df_game['Critic_Score'].dropna()
df_score
```

```python
df_score = df_score.to_numpy()
```

```python
stat, p = st.shapiro(df_score)

print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Вероятно нормальное распределение')
else:
    print('Вероятно не нормальное распределение')
```

Распределение вероятно не нормальное, для проверки гипотезы используем Тест Манна-Уитни

```python
df_game['Critic_Score']
```

```python
all = df_game['Critic_Score'].dropna()
all
```

```python
sport = df_game[df_game['Genre'] == 'Sports']['Critic_Score'].dropna()
sport
```

```python
stat, p = st.mannwhitneyu(all, sport)

print(f"Статистика = {stat:.5f}, p = {p:.5f}")

if p > 0.05:
    print('Не отклоняем нулевую гипотезу, распределения, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, распределения, вероятно, различаются')
```

```python
print(sport.mean(), all.mean())
```

Вывод: Распределение отличается, возможно, критикам больше нравятся спортивные игры

Проверка гипотезы: Критикам нравятся больше игры на PC или на PS4?

*   Н0 - Оценки критиков игр на PC и PS4 не отличаются
*   Н1 - Оценки критиков игр на PC и PS4 статистически отличаются


```python
df_pc = df_game[df_game['Platform'] == 'PC']['Critic_Score'].dropna()
```

```python
df_ps4 = df_game[df_game['Platform'] == 'PS4']['Critic_Score'].dropna()
```

```python
stat, p = st.mannwhitneyu(df_pc, df_ps4)

print(f"Статистика = {stat:.5f}, p = {p:.5f}")

if p > 0.05:
    print('Не отклоняем нулевую гипотезу, распределения, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, распределения, вероятно, различаются')
```

```python
print(df_pc.mean(), df_ps4.mean())
```

Вывод: Оценки критиков игр на PC и PS4 статистически отличаются, возможно критикам больше нравятся игры на PC

Проверка гипотезы: Критикам больше нравятся стрелялки или стратегии?

*   Н0 - Распределение данных статистически не отличается
*   Н1 - Распределение данных статистически отличается



```python
df_shooter = df_game[df_game['Genre'] == 'Shooter']['Critic_Score'].dropna()
```

```python
df_strategy = df_game[df_game['Genre'] == 'Strategy']['Critic_Score'].dropna()
```

```python
stat, p = st.mannwhitneyu(df_shooter, df_strategy)

print(f"Статистика = {stat:.5f}, p = {p:.5f}")

if p > 0.05:
    print('Не отклоняем нулевую гипотезу, распределения, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, распределения, вероятно, различаются')
```

```python
print(df_shooter.mean(), df_strategy.mean())
```

Вывод: Оценки критиков игр шутеров и стратегий статистически отличаются, возможно критикам больше нравятся стратегии

```python
df_message = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/stat_case_study/spam.csv')
df_message
```

```python
df_message['Message'] = df_message['Message'].str.lower()
df_message
```

```python
df_message['Message'] = df_message['Message'].replace(to_replace ='[\W_]+', value = ' ', regex = True)
df_message['Message']
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords_set = set(stopwords.words('english'))
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('omw-1.4')
nltk.download('wordnet')

wordnet_lemmatizer = WordNetLemmatizer()
```

```python
def stop_lem(row):
  str_ = ''
  for word in row['Message'].split():
    if word not in stopwords_set:
      word = wordnet_lemmatizer.lemmatize(word)
      str_ += word + ' '
  return str_
```

```python
df_message.apply(stop_lem, axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df_message['Message'])
names = tfidf.get_feature_names_out()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
```

```python
tfidf_matrix
```

```python
df_message['Category'] = df_message['Category'].replace('ham', 1)
df_message['Category'] = df_message['Category'].replace('spam', 0)
```

```python
df_message
```

```python
y = df_message['Category'].to_numpy()
y
```
