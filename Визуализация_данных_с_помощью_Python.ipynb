```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
playstore = pd.read_csv('googleplaystore.csv')
```

```python
playstore.head()
```

```python
app_category = playstore[['Category', 'App']].groupby('Category').count()
app_category
```

```python
app_category.loc[['FAMILY','GAME','TOOLS','MEDICAL','BUSINESS']].sort_values(by='App', ascending=False).plot(kind='bar',
                                                                                                             title='Топ 5 категорий, в которых больше всего приложений',
                                                                                                             xlabel='Категория приложений',
                                                                                                             ylabel='Количество приложений')
plt.legend().set_visible(False)
```

Вывод:На графике мы можем увидеть самые популярные категории приложений, в которых их представлено больше всего

```python
playstore['Genres'].unique()
```

```python
playstore['Genres'] = playstore['Genres'].replace('February 11, 2018', 'NaN')
playstore['Genres'].unique()
```

```python
top5_genres = playstore[['Rating','Genres']].groupby('Genres').mean().sort_values('Rating', ascending=False).head(6)
top5_genres.drop(index=top5_genres.index [0], axis= 0 , inplace= True )
```

```python
top5_genres
```

```python
top5_genres.plot(kind='bar', title='Топ 5 жанров приложений по рейтингам',
                             xlabel='Жанры',
                             ylabel='Рейтинг')
for i, val in enumerate(top5_genres.values):
    plt.text(i, val, val, horizontalalignment='center', verticalalignment='bottom', fontdict={'size':14})
plt.legend().set_visible(False)
```

Вывод: на графике представлены жанры, в которых больше всего приложений с высоким рейтингом

```python
installs = playstore[['Installs', 'App']].groupby('Installs').count().sort_values('Installs',ascending=False).head(6)
```

```python
installs.reset_index().replace('Free', 'NaN')
```

```python
installs.drop(index=installs.index [0], axis= 0 , inplace= True )
```

```python
installs
```

```python
installs.plot(kind='pie',subplots=True)
plt.legend(bbox_to_anchor=(0.6, .5, 0.5, 0.5))
```

Вывод: на графике можно увидеть топ 5 самых больших скачиваний по приложениям
