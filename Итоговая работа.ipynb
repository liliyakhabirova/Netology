Проверяем нормальность распределения данных показателя 'Среднее количество часов на рабочем месте в месяц' тестами Шапиро-Уилка и Андерсона-Дарлинга:

*  Н0 - Данные распределены по нормальному закону.
*  Н1 - Данные не распределены по нормальному закону.





```python
df[['average_montly_hours', 'salary']]
```

```python
data = df['average_montly_hours']
data
```

```python
data.plot(kind='hist', title = 'Распределение данных среднего количества часов на рабочем месте в месяц' ,bins=20)
```

```python
stat, p = st.shapiro(data)

print('stat=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
  print('Вероятно нормальное распределение')
else:
  print('Вероятно не нормальное распределение')
```

```python
result = st.anderson(data)
```

```python
print('stat=%.3f' % (result.statistic))
for i in range(len(result.critical_values)):
    sl, cv = result.significance_level[i], result.critical_values[i]
    if result.statistic < cv:
        print('Вероятно нормальное на уровне %.1f%%' % (sl))
    else:
        print('Вероятно не нормальное на уровне %.1f%%' % (sl))
```

Проверяем гипотезу о то что, сотрудники с высоким окладом
проводят на работе больше времени, чем сотрудники с низким
окладом. Проверяем непараметрическим тестом Манна-Уитни, так как распределение данных не подчиняется нормальному закону

*   Н0 - Распределение данных среднего количество часов на рабочем месте в месяц у сторудников с высокой зарплатой и распределение данных среднего количества часов на рабочем месте в месяц у сотрудников с низкой зарплатой не различаются
*   Н1 - Распределение данных среднего количество часов на рабочем месте в месяц у сторудников с высокой зарплатой и распределение данных среднего количества часов на рабочем месте в месяц у сотрудников с низкой зарплатой различаются

```python
df_high = df[df['salary'] == 'high']['average_montly_hours']
df_high
```

```python
df_high.count()
```

```python
df_low = df[df['salary'] == 'low']['average_montly_hours']
df_low
```

```python
df_low.count()
```

```python
stat, p = st.mannwhitneyu(df_high, df_low)

print(f"Статистика = {stat:.5f}, p = {p:.5f}")

if p > 0.05:
    print('Не отклоняем нулевую гипотезу, распределения, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, распределения, вероятно, различаются')
```

Вывод: Согласно непараметрическому тесту Манна-Уитни нельзя сказать что сотрудники с высоким окладом проводят на рабочем месте статистически значимое большее количество часов рабочего времени.

```python
data_class = df['left']
data_class
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_model, data_class, random_state=42)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(X_train, y_train)
```

```python
lda.predict(X_test)
```

```python
result = pd.DataFrame([y_test, lda.predict(X_test)]).T
result
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, lda.predict(X_test))
```
